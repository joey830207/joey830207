```python
import numpy as np
import matplotlib.pyplot as plt

# Load the npz file
data = np.load(r"C:\Users\Joey\Desktop\Rabi_AWG_2.6300GHz_-6.0dBm_400x_frames_222.npz")

# Get data from the file
arr_meas = data['frames_meas']
arr_ctrl = data['frames_ctrl']
#print(np.shape(arr_meas))

# sum up the elements in each 2D array individually
meas_1d_sum = np.sum(np.sum(arr_meas, axis=1), axis=1)
meas_1d_sum = meas_1d_sum.astype(np.int64)
ctrl_1d_sum = np.sum(np.sum(arr_ctrl, axis=1), axis=1)
ctrl_1d_sum = ctrl_1d_sum.astype(np.int64)
result = np.subtract(ctrl_1d_sum, meas_1d_sum)
print(meas_1d_sum)
print(ctrl_1d_sum)
print(result)

# insert time array
time_arr = np.arange(0., 1000., 5.)
data_f = np.transpose(np.stack((time_arr, meas_1d_sum, ctrl_1d_sum, result)))

# specify the file path to your desktop directory
file_path = r'C:\Users\Joey\Desktop\rabi1.csv'

# save the 2D array to a CSV file
np.savetxt(file_path, data_f, delimiter=',')

# # reshape the 3D array into a 2D array
# arr_2d = arr.reshape(-1, arr.shape[-1])

# # specify the file path to your desktop directory
# file_path = r'C:\Users\Joey\Desktop\meas.csv'

# # save the 2D array to a CSV file
# np.savetxt(file_path, arr_2d, delimiter=',')



# # Print the keys of the npz file
# print("Keys in the npz file:", data.files)

# # Loop through the keys and print the corresponding values
# for key in data.files:
#     print(key, ":")
#     print(data[key])
```
